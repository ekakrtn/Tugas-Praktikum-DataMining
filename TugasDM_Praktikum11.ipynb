```python
!pip install pandas statsmodels
import pandas as pd
import statsmodels.api as sm
df = pd.read_csv('Data Penjualan Toko Buku.csv',sep = ',')
df.rename(columns={'jumlah':'Jumlah item'}, inplace=True)
df.rename(columns={'total':'Jumlah Transaksi'}, inplace=True)
df
```

```python
df = df.dropna()
```

```python
# Membersihkan data
df['Jumlah item'] = pd.to_numeric(df['Jumlah item'], errors='coerce')  # Pastikan kolom numerik
df['Jumlah Transaksi'] = pd.to_numeric(df['Jumlah Transaksi'], errors='coerce')  # Pastikan kolom numerik
df = df.dropna()
```

```python
#Statistik deskriptif
df.describe()
```

```python
# Menentukan variabel independen dan dependen
X = df['Jumlah item']
y = df['Jumlah Transaksi']
```

```python
# Menambahkan konstanta
X = sm.add_constant(X)
```

```python
# Membuat model OLS
model = sm.OLS(y, X).fit()
```

```python
# Menampilkan hasil
print(model.summary())
```
